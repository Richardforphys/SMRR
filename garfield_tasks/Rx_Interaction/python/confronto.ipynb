```python
c_Ar = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/counts_Ar.npy")
e_Ar = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/Ecounts_Ar.npy")
c_Ar9010 = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/counts_Ar9010.npy")
e_Ar9010 = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/Ecounts_Ar9010.npy")
c_Kr = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/counts_Kr.npy")
e_Kr = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/Ecounts_Kr.npy")
c_Xe = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/counts_Xe.npy")
e_Xe = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/Ecounts_Xe.npy")
c_XeSF6 = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/counts_XeSF6.npy")
e_XeSF6 = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/.npy/Ecounts_XeSF6.npy")
```

```python
cer_Ar = np.sqrt(c_Ar*5e4)/5e4
cer_Ar9010 = np.sqrt(c_Ar9010*5e4)/5e4
cer_Kr = np.sqrt(c_Kr*5e4)/5e4
cer_Xe = np.sqrt(c_Xe*5e4)/5e4
cer_XeSF6 = np.sqrt(c_XeSF6*5e4)/5e4
```

```python
fig, axs = plt.subplots(1, 4, figsize=(15,5), sharey=True)
axs[2].errorbar(
    e_Kr,
    c_Kr,
    yerr=cer_Kr,
    fmt="--",
    color="red",
    capsize=4
)
axs[3].errorbar(
    e_Ar,
    c_Ar,
    yerr=cer_Ar,
    fmt="--",
    color="navy",
    capsize=4,
    label='(70/30)'
)
axs[3].errorbar(
    e_Ar9010,
    c_Ar9010,
    yerr=cer_Ar9010,
    fmt="--",
    color="blue",
    capsize=4,
    label='(90/10)'
)
axs[0].errorbar(
    e_Xe,
    c_Xe,
    yerr=cer_Xe,
    fmt="--",
    color="purple",
    capsize=4
)
axs[1].errorbar(
    e_XeSF6,
    c_XeSF6,
    yerr=cer_XeSF6,
    fmt="--",
    color="pink",
    capsize=4
)

axs[0].set_ylabel(r"$\mu$")
axs[0].set_title(r"XeCO$_2$")
axs[1].set_title(r"XeCO$_2$")
axs[2].set_title(r"KrCO$_2$")
axs[3].set_title(r"ArCO$_2$")
for i in range(4):
    axs[i].set_xlabel("Energia [keV]") 
    axs[i].grid()
    axs[i].legend()
plt.tight_layout()
plt.savefig(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/PNG_utili/mu.pdf", dpi=300)
```
