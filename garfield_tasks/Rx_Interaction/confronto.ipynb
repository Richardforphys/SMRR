```python
c_Ar = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/counts_Ar.npy")
e_Ar = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/Ecounts_Ar.npy")
c_Kr = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/counts_Kr.npy")
e_Kr = np.load(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/Ecounts_Kr.npy")
```

```python
cer_Ar = np.sqrt(c_Ar*5e4)/5e4
cer_Kr = np.sqrt(c_Kr*5e4)/5e4

plt.figure(figsize=(15,5))
plt.errorbar(
    e_Kr,
    c_Kr,
    yerr=cer_Kr,
    fmt=".",
    color="red",
    alpha=0.5,
    capsize=4
)
plt.errorbar(
    e_Ar,
    c_Ar,
    yerr=cer_Ar,
    fmt=".",
    color="navy",
    alpha=0.5,
    capsize=4
)

plt.xlabel("Energia [keV]")
plt.ylabel(r"$\lambda$")
plt.grid()
plt.tight_layout()
plt.savefig(r"/home/ricca/SMRR/garfield_tasks/Rx_Interaction/PNG_utili/LAmbdasd.pdf", dpi=300)
```
